```python
left_diabetic = df.loc[(df['D'] ==1) & (df['Left-Diagnostic Keywords'].apply(lambda x: 1 if ("diabetic" in x or "proliferative" in x) else 0) == 1)]['Left-Fundus'].values
right_diabetic = df.loc[(df['D'] ==1) & (df['Right-Diagnostic Keywords'].apply(lambda x: 1 if ("diabetic" in x or "proliferative" in x) else 0) == 1)]['Right-Fundus'].values
left_diabetic.shape, right_diabetic.shape
#
```

```python
left_normal = df.loc[(df['D'] == 0) & (df['Left-Diagnostic Keywords'].apply(lambda x: 1 if "normal" in x else 0) == 1)]['Left-Fundus'].values
right_normal = df.loc[(df['D'] == 0) & (df['Right-Diagnostic Keywords'].apply(lambda x: 1 if "normal" in x else 0) == 1)]['Right-Fundus'].values
left_normal.shape, right_normal.shape
```

```python
diabetic = np.concatenate((left_diabetic,right_diabetic),axis=0)
normal = np.concatenate((left_normal, right_normal), axis=0)
diabetic.shape, normal.shape
#
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32
NO_EPOCHS = 50
NUM_CLASSES = 2
DATA_FOLDER = "ODIR-5K/ODIR-5K/Training Images"
```

```python
dataset = []
def create_dataset(image_category,label):
    
    for img in tqdm(image_category):
        img_path = os.path.join(DATA_FOLDER,img)
        image = cv.imread(img_path,cv.IMREAD_COLOR)
        image = cv.resize(image,(IMG_SIZE,IMG_SIZE))
        
        dataset.append([np.array(image),np.array(label)])
    random.shuffle(dataset)
    return dataset
```

```python
dataset = create_dataset(diabetic,1)
```

```python
dataset = create_dataset(normal,0)
```

```python
plt.figure(figsize=(12,7))
for i in range(10):
    sample = random.choice(range(len(dataset)))
    image = dataset[sample][0]
    category = dataset[sample][1]
    if category== 0:
        label = "Normal"
    else:
        label = "Diabetic"
    plt.subplot(2,5,i+1)
    plt.imshow(image)
    plt.xlabel(label)
plt.tight_layout()    
```

```python
dataset[0]
```

```python
x = np.array([i[0] for i in dataset]).reshape(-1,IMG_SIZE,IMG_SIZE,3)
y = np.array([i[1] for i in dataset])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
model = Sequential()
model.add(ResNet50(include_top=False, pooling='max', weights='imagenet'))
model.add(Dense(1, activation='softmax'))
# ResNet-50 model is already trained, should not be trained
model.layers[0].trainable = True
```

```python
opt = tfa.optimizers.LazyAdam()
loss = tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.025)
model.compile(optimizer='sgd', loss=loss, metrics=['accuracy'])
```

```python
y_test.shape
```

```python
model.summary()
```

```python
train_model = model.fit(x_train, y_train,
                  batch_size=BATCH_SIZE,
                  epochs=NO_EPOCHS,
                  verbose=1,
                  validation_data=(x_test, y_test))
```

```python
plot_model(model, to_file='model.png')
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```
